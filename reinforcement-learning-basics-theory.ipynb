## 1. Reinforcement Learning Introduction

Reinforcement learning (RL) algorithms allow to find the best way for an agent (a subject
with several possible actions) to behave in a given environment.

The environment defines what are the different possible actions by the agent. This agent must
be able to be in different situations and his situations will be called the agents different states.
For each state of the agent, a multitude of choices are given by the environment, the agent
performs one of these actions. Knowing this, the environment defines what is the new state of
the agent.

In addition to giving a new state, the environment also gives a quantitative value called
reward. This value is very important and is used in the different RL algorithms.
This value defines whether the action taken by the agent in his state is a good decision or not.
The agent and the environment interact together, this interaction is described in following image.
For example, in a video game, the agent would be the player who makes decisions based on
the state of his character in the game. The environment represents the game itself and the map

## 2. Markov Decision Process

The problem must be solved by a machine learning method, the theory of situation resolution
by machine learning algorithms is very elaborate.

A Markovian Decision-making Process (MDP) is a stochastic model where an agent makes
decisions and where the results of his actions are random.

A markovian decision-making process is a discrete stochastic control process. At each step,
the process is in a certain state S, and the agent chooses an action A.

The probability that the process will reach the state is determined by the chosen action. More
precisely, it is described by a state transition function. Thus, the next state is dependent on the
current state and the action selected by the decision-maker.
However, for a state and an action, the next state is independent of previous actions and
states.

Some machine learning algorithms cannot be use if the condition of MDP is not respected.
Because of the complexity of the environment to study, algorithms that does not need to
respect a MDP are more interesting.
The following algorithms can be used without being in a MDP.

The second model based algorithm is the value iteration, it computes the optimal state-value

It repeatedly updates the Q(s, a) and V(s) values until they converge to the same

## 3.3 Model based uses

To use a model based algorithm some prerequisites are needed, first of all, the environment
must be a deterministic environment which means that the couple of a given state and a given
action will always provide the same next state and receive the same reward, an easier way to
describe a deterministic environment is to imagine an environment without any part of
randomness.
Secondly the algorithm has to know the reward function, thirdly it need to know
the transition function.

Both policy iteration and value iteration are guaranteed to converge to an optimal policy but
the main difference between them is the computating power necessary.

Policy-iteration takes considerably fewer number of iterations to converge but on the other
hand each iteration is more computationally expensive so for a size limited situation the
policy-iteration is more efficient while in a more complex situation the value-iteration will
use less computation because in the case of many iterations the policy iteration will need an
amazing amount of computating power compare to the value-iteration method.

Model based algorithm should be used in deterministic environment and knowing both the
transition function and the reward, only use it for limited environment because it will need to
much computating power for a continuous and spacious environment.

### 4.1.2 The Learning Rate

The learning rate is a number between 0 and 1, it describes the ratio between the importance
of the new value in the Q table and the old one. 

A learning rate of 1 will totally erase the old
value while a learning rate of 0 will always keep the old value, in that case the algorithm
becomes totally useless because it doesnâ€™t learn anymore.
A learning rate of 0.1 is often used to have a good convergence and a good ration between the
old and the new Q value.

### 4.1.3 The Discount Factor

The discount factor is also a number between 0 and 1. It describe the capacity of the algorithm
to evaluate further reward, a discount factor of 1 will not take in consideration the number of
actions necessary to get a reward, it will always look for the biggest reward wherever it can be. 
This can lead to an infinite program. 

A discount factor of 1 will make the program blind, it will only look for instant reward which doesnâ€™t need any more actions.

### 4.1.4 The Reward R

The reward is the amount given by the environment to update the Q-table. It can be a reward
(positive points) or a punishment (negative points).

The reward function can be changed to optimize different criteria. It is important to understand that the optimal policy will depend of
the reward function. It is very important to define with care the reward function otherwise the

These different parameters influence the results and training time, they need to be optimize

## 4.2 Exploration/exploitation Trade-Off

To get a good convergence we need to get the maximum of iteration for every cell of the Q-
table, we need to take most of the time the best decision otherwise our algorithm will never
have an optimal policy, but we also need to explore all situations to make sure we are not
missing a better one.
This dilemma is called the exploration and exploitation trade-off. There
are different approaches:

* 1. The greedy approach

The greedy approach consists of always taking the best option (highest Q value) but this
approach is not really useful because the machine is not going to learn because from the
moment it finds a Q-value higher than all the others, it will always select the same action.
This approach provides 100% of exploitation and 0% of exploration.

* 2. The Random approach

The random approach consists of taking very often a random action, it provides either
exploration and exploitation, but this approach is using too much randomness (75%
exploration/25% exploitation). This is not the optimal solution, but this is already a good
exploration/exploitation Trade-Off.


* 3. The Îµ-greedy approach

The Îµ-greedy approach is a combination of the two previous methods, it takes most of the
time the best option (highest Q value) but sometimes it takes a random action excluding the
best one. It provides a very good exploration-exploitation trade-off (75% exploitation/25%
exploration).
This approach is the most used and it has a very good ratio between simplicity and efficiency.


* 4. The Boltzmann approach

The Boltzmann approach is the last method we can use, just as the Îµ-greedy approach it takes
most of the time the best option and sometimes takes a random action excluding the best one
but the difference between these two approaches is that for the Boltzmann approach the
probability to be taken for each action depends of the Q-value of itself.

The Boltzmann approach is more efficient than the Îµ-greedy approach and converge faster, on
the other hand it is a method mathematically more complex and harder to implement.

This difference of complexity can be explained by the function to use for each approach, for
the Îµ-greedy approach the function selects a random number between 0 and 1 and if it is
between 0 and 0.75 the best option is selected otherwise another random action is selected
whereas the Boltzmann approach uses the Boltzmann equation

### 4.4 Model free uses

Unlike the model based model, a model free model does not need to know the transition
function and the reward, they will be managed by the environment.

Q-learning and SARSA algorithms can be used in stochastic environment (an environment
who include a part of randomness: do the same action in the same state will note always
involve the same reward), this is the main asset of model free algorithms.


Both algorithms converge to an optimal policy but SARSA being an on-policy method will
follow a safe road while Q-learning will find the fastest way even if it means losing more
often.

To know which algorithm use, it is important to evaluate if a loss is something to avoid during
the training or not, for example during an offline test for a game, a loose is not costly so it is
better to use the Q-learning whereas if an expensive robot is used during the training, you
should better use the SARSA algorithm.
